```python
import pdfplumber
import pandas as pd
from datetime import datetime
pdf = pdfplumber.open(r"C:\Users\esfita-user\Downloads\canara_epassbook_2023-11-07 19_19_23.891379.pdf")
date = []
desc = []
chq_no = []
amount = []
debit = []
credit = []
balance = []
i = ""
j = 0
for page in range(len(pdf.pages)):
    bank = pdf.pages[page].extract_text()
    bank = bank.split("\n")
    for line in range(len(bank)):
        trans = bank[line].split()
        if len(trans) > 2:
            if len(str(trans[0]).split("-")) == 3 and len(str(trans[-1]).split(".")) == 2:
                print(trans)
                trans1 = bank[line-1]
                print(trans1)
                trans2 = bank[line-2]
                print(trans2)
                trans3 = bank[line-3]
                print(trans3)
                date.append(trans[0])
                if len(trans) == 3:
                    desc.append(trans3 + trans2 + trans1)
                else:
                    desc.append(trans3+trans2)
                chq_no.append(i)
                amount.append(trans[-2])
                balance.append(trans[-1])
print(date)
print(desc)
print(chq_no)
print(amount)
print(balance)
df = pd.DataFrame([date,desc,chq_no,amount,balance]).T
df[4] = df[4].str.replace(",","").apply(lambda x : float(x))
df["debit"] = 0
df["credit"] = 0
for data in range(len(df)):
    if df.iloc[data-1,4] > df.iloc[data,4]:
        df.iloc[data,-2] = df.iloc[data,3]
    else:
        df.iloc[data,-1] = df.iloc[data,3]
new_col = [0,1,2,"debit","credit",4]
df = df[new_col]
print(df)
df.rename(columns = {0 : "Date",1 : "Description",2 : "Chq_no","debit" : "Debit","credit" : "Credit",4 : "Balance"},inplace = True)
print(df)
df.to_excel("canara bank 136pages.xlsx",index = False)
```
